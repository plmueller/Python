```python
import pandas as pd
```

```python
mpg = pd.read_csv('/Users/patrickmueller/Downloads/seaborn-data-master/mpg.csv')
```

```python
import seaborn as sns
sns.set_theme()
mpg = sns.load_dataset('mpg')
```

```python
#Create a visual showing the relationship between horsepower and fuel economy

sns.relplot(data=mpg, x = 'mpg', y = 'horsepower')
```

## The chart above shows that the more horsepower you have, the lower the mpg is, with the best fuel economy having less horsepower

```python
#Create a table summarizing fuel economy by country of origin

sns.catplot(data=mpg, x = 'origin', y = 'mpg', kind = 'bar', errorbar = None)
```

## The chart above shows the average fuel economy based on the origin. USA has the lowest avg mpg (20) where Japan has the best avg mpg (31)

```python
#Create a visual showing how fuel economy has changed over time

sns.catplot(data=mpg, x = 'model_year', y = 'mpg', kind = 'bar', errorbar = None)
```

## The chart above shows the avg mpg per model year. It demonstrates a steady increase over the 12 years starting at around 17 mpg improving to 32 mpg. That's almost double in a 12 year time period.

```python
# Show the top 5 cars with the best fuel economy

top5cars = mpg.groupby('name').mpg.mean().sort_values(ascending=False).head()
top5carsdf = top5cars.to_frame().reset_index()
sns.catplot(data=top5carsdf, x = 'mpg', y = 'name', kind = 'bar', errorbar = None)
```

## The chart above shows the top 5 cars relative to mpg with the very best being the Mazda GLC

```python
# Show the relationship between the weight and fuel economy

sns.relplot(data=mpg, x = 'mpg', y = 'weight')
```

## The chart above shows that as weight decreases in a given vehicle, the fuel economy will increase. This would be expected as the less a vehicle weighs, the less energy(gas) is needed to power the vehicle.
